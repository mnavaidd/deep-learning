# Binarize The Images

```
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# All images will be rescaled by 1./255
train_datagen = ImageDataGenerator(rescale=1./255) # set validation split
val_datagen = ImageDataGenerator(rescale=1./255)
```

# Resize The Images

```
train_generator = train_datagen.flow_from_directory(
        train_dir,
        batch_size=32,
        target_size=(256, 256), #image resize
        class_mode='categorical'
        )

validation_generator = val_datagen.flow_from_directory(
        validation_dir,
        batch_size=32,
        target_size=(256, 256),
        class_mode='categorical'
        )

print(train_generator.class_indices.keys())
print(train_generator.class_indices.values())
print(validation_generator.class_indices.keys())
```

# Classes

```
print(train_generator.class_indices.keys())
print(validation_generator.class_indices.keys())
```

# Create Model

```
from tensorflow.keras import models
from tensorflow.keras import layers

model = models.Sequential([
    layers.Flatten(input_shape=(256,256,3)),
    layers.Dense(512, activation='relu'),
    layers.Dense(256, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(39, activation='softmax')
])
```

# Compile The Model

```
model.compile(optimizer='rmsprop',
                       loss='categorical_crossentropy',
                       metrics=['acc'])
```

```
model.summary()
```

# Train The Model

```
model.fit(train_generator,
    steps_per_epoch=100,
    epochs=50,
    validation_data=validation_generator,
    validation_steps=50)
```

```
opt = tensorflow.keras.optimizers.Adam(0.1)
ckpt = tensorflow.train.Checkpoint(step=tensorflow.Variable(1), optimizer=opt, net=model)
manager = tensorflow.train.CheckpointManager(ckpt, './tf_ckpts', max_to_keep=3)
```

```
def train_and_checkpoint(net, manager):
  ckpt.restore(manager.latest_checkpoint)
  if manager.latest_checkpoint:
    print("Restored from {}".format(manager.latest_checkpoint))
  else:
    print("Initializing from scratch.")
```
